##### 2) CLEAN DATA

##### Since we have different time perios in each dataset we need to transform it into same time unit. As we have, daily, weekly, montly and quarterly we are going to take the month as the reference

##### 2.1 Transform S&P into monthly observations

```python
monthly_data_list = []
month_list = sorted(sp_500.index.month.unique())
year_list = sp_500.index.year.unique()

for year in year_list:
    for month in month_list:
        last_day_data = sp_500[(sp_500.index.year == year) & (sp_500.index.month == month)].tail(1)
        monthly_data_list.append(last_day_data)
        
sp500_monthly = pd.concat(monthly_data_list)

sp500_monthly.head()
```

```python
# Now we want to keep only the close which is the value of the sp500 at the end of the session for each year. Which is the same of beginning of next year

sp500_monthly_performance = sp500_monthly[["Close"]] / sp500_monthly[["Close"]].shift(1) * 100 - 100

sp500_monthly_performance.head()

# Now we need to restart index to 01-01-YYYY

sp500_monthly_performance.index = pd.to_datetime(sp500_monthly_performance.index.strftime("%Y-%m")) # Reset index to 01-01-yyyy

sp500_monthly_performance.tail(5)

sp500_monthly_performance = sp500_monthly_performance.rename(columns={"Close": "close_sp500"})

sp500_monthly_performance.head()
```

##### 2.2 Transform Bonds into monthly observations

```python
monthly_data_list = []
month_list = sorted(bonds.index.month.unique())

for year in year_list:
    for month in month_list:
        last_day_data = bonds[(bonds.index.year == year) & (bonds.index.month == month)].tail(1)
        monthly_data_list.append(last_day_data)
        
bonds_monthly = pd.concat(monthly_data_list)

bonds_monthly.head()
```

```python
bonds_monthly_performance = bonds_monthly[["Close"]] / bonds_monthly[["Close"]].shift(1) * 100 - 100

bonds_monthly_performance.head()

# Now we need to restart index to 01-01-YYYY

bonds_monthly_performance.index = pd.to_datetime(bonds_monthly_performance.index.strftime("%Y-%m")) # Reset index to 01-01-yyyy

bonds_monthly_performance.head(5)

bonds_monthly_performance = bonds_monthly_performance.rename(columns={"Close": "close_bonds"})
```

##### 2.3 Transform Cash into monthly observations

```python
monthly_data_list = []
month_list = sorted(cash.index.month.unique())

for year in year_list:
    for month in month_list:
        last_day_data = cash[(cash.index.year == year) & (cash.index.month == month)].tail(1)
        monthly_data_list.append(last_day_data)
        
cash_monthly = pd.concat(monthly_data_list)

cash_monthly.head()

cash_monthly_performance = cash_monthly[["Close"]] / cash_monthly[["Close"]].shift(1) * 100 - 100

cash_monthly_performance.head()

# Now we need to restart index to 01-01-YYYY

cash_monthly_performance.index = pd.to_datetime(cash_monthly_performance.index.strftime("%Y-%m")) # Reset index to 01-01-yyyy

cash_monthly_performance.tail(5)

cash_monthly_performance = cash_monthly_performance.rename(columns={"Close": "close_cash"})
```

##### 2.4 Transform Gold into monthly observations

```python
monthly_data_list = []
month_list = sorted(gold.index.month.unique())

for year in year_list:
    for month in month_list:
        last_day_data = gold[(gold.index.year == year) & (gold.index.month == month)].tail(1)
        monthly_data_list.append(last_day_data)
        
gold_monthly = pd.concat(monthly_data_list)

gold_monthly.head()

gold_monthly_performance = gold_monthly[["Close"]] / gold_monthly[["Close"]].shift(1) * 100 - 100

gold_monthly_performance.head()

# Now we need to restart index to 01-01-YYYY

gold_monthly_performance.index = pd.to_datetime(gold_monthly_performance.index.strftime("%Y-%m")) # Reset index to 01-01-yyyy

gold_monthly_performance = gold_monthly_performance.rename(columns={"Close": "close_gold"})

gold_monthly_performance.head(5)
```

```python
merge_df_assets = pd.merge(sp500_monthly_performance, bonds_monthly_performance, left_index = True, right_index = True)

merge_df_assets.head()
```

```python
merge_df_assets1 = pd.merge(merge_df_assets, cash_monthly_performance, left_index = True, right_index = True)
```

```python
merge_df_assets2 = pd.merge(merge_df_assets1, gold_monthly_performance, left_index = True, right_index = True)

merge_df_assets2.head()
```

```python
merge_df = pd.merge(merge_df_assets2, cpi_monthly, left_index = True, right_index = True)

merge_df.head()
```

```python
merge_df1 = pd.merge(merge_df, ir, left_index = True, right_index = True)

merge_df1.head()
```

```python
merge_df2 = pd.merge(merge_df1, gdp, left_index = True, right_index = True)

merge_df2.head()
```

```python
correlation_matrix = merge_df2.corr()
correlation_matrix = correlation_matrix.round(2)
display(correlation_matrix)
```
